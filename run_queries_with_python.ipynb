## Searching for אִגֶּרֶת in the MT

In this notebook we explore how to run a Search query with Python. We search for clauses which contain the word אִגֶּרֶת "letter". First, we load some packages and the BHSA dataset.

```python
import pandas as pd
import seaborn as sns

from tf.app import use
BHS = use('etcbc/bhsa', version="2021", hoist=globals())
```

Here you see the query and how it is executed. With .show() you can see the results. Here we show only the first result with some extra features.

```python
query = '''
clause
   word lex=>GRT/
'''

letter = BHS.search(query)

BHS.show(letter, start=1, end=1, extraFeatures={'nu', 'ps'})
```

Now we print all the results.

```python
BHS.show(letter, extraFeatures={'nu', 'ps'})
```
